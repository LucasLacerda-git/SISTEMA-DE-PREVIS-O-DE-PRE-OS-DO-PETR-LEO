````python
# IMPORTS
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import warnings
import random

warnings.filterwarnings('ignore')

# FIXANDO A SEED PARA REPRODUZIBILIDADE
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)
````

##Análise Explorátoria

````python
df = pd.read_csv("/content/train_data.csv")
df.rename(columns={"Petrol (USD)": "Petrol"}, inplace=True)
df['Date'] = pd.to_datetime(df['Date'])

print("RESUMO DOS DADOS:")
print(f"Período: {df['Date'].dt.year.min()}-{df['Date'].dt.year.max()}")
print(f"Registros: {len(df):,}")
print(f"Preço Médio: ${df['Petrol'].mean():.2f}")
print(f"Volatilidade: ${df['Petrol'].std():.2f}")
````

````python
#Análise de Sazionalidade
df['Ano'] = df['Date'].dt.year
df['Mes'] = df['Date'].dt.month
preco_por_ano = df.groupby('Ano')['Petrol'].mean()
print(f"\n Tendência Anual:")
for ano, preco in preco_por_ano.items():
    print(f"   {ano}: ${preco:.2f}")
````

-Dados mostram alta volatilidade com tendências anuais claras

## Tratamento de Dados

````python
df_clean = df[(df['Petrol'] > 50) & (df['Petrol'] < 200)].copy()

print(f" Dados após limpeza: {len(df_clean):,} registros")
print(f" Estatísticas finais:")
print(f"   Média: ${df_clean['Petrol'].mean():.2f}")
print(f"   Min: ${df_clean['Petrol'].min():.2f}")
print(f"   Max: ${df_clean['Petrol'].max():.2f}")
# Removeedo valores impossíveis mantendo volatilidade real
````

- # Engenharia: Preservação da volatilidade real do mercado


````python
# Gráfico de linha temporal
fig = px.line(df_clean, x='Date', y='Petrol',
              title='<b>Evolução Histórica dos Preços do Petróleo</b>',
              labels={'Petrol': 'Preço (USD/barril)', 'Date': 'Data'})
fig.update_layout(template='plotly_white', title_x=0.5)
fig.show()
````

````python
# Heatmap de sazonalidade mensal
pivot_table = df_clean.pivot_table(values='Petrol', index='Mes', columns='Ano', aggfunc='mean')
fig = px.imshow(pivot_table, title='<b>Padrões Sazonais por Mês e Ano</b>')
fig.update_layout(template='plotly_white', title_x=0.5)
fig.show()
````

- Identificação clara de tendências e sazonalidades

##Preparação

````python
scaler = MinMaxScaler(feature_range=(0, 1))
df_clean['Petrol_Scaled'] = scaler.fit_transform(df_clean[['Petrol']])
````

````python
# Divisão temporal (70% treino, 30% teste)
train_size = int(len(df_clean) * 0.7)
train_data = df_clean['Petrol_Scaled'].iloc[:train_size].values
test_data = df_clean['Petrol_Scaled'].iloc[train_size:].values

print(f" Divisão dos Dados:")
print(f"   Treino: {train_size} registros (70%)")
print(f"   Teste: {len(test_data)} registros (30%)")
````

##CONSTRUÇÃO DE SEQUÊNCIAS TEMPORAIS

````python
def criar_sequencias(dados, time_steps=60):
    X, y = [], []
    for i in range(time_steps, len(dados)):
        X.append(dados[i-time_steps:i])
        y.append(dados[i])
    return np.array(X), np.array(y)

TIME_STEPS = 60
X_train, y_train = criar_sequencias(train_data, TIME_STEPS)
X_test, y_test = criar_sequencias(test_data, TIME_STEPS)
````

````python
# Reformatar para LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
````

````python
print(f" Formatos das Sequências:")
print(f"   X_train: {X_train.shape}")
print(f"   X_test: {X_test.shape}")
#Modelagem - 60 dias históricos para prever o próximo dia
````

##ARQUITETURA DA REDE NEURAL

````python
modelo = Sequential([
    LSTM(100, return_sequences=True, input_shape=(TIME_STEPS, 1)),
    Dropout(0.2),
    LSTM(100, return_sequences=True),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(1)
])

modelo.compile(optimizer='adam', loss='mse', metrics=['mae'])

print(" Arquitetura do Modelo:")
print("   3 camadas LSTM com Dropout")
print("   Otimizador: Adam")
print("   Loss: Mean Squared Error")
````

##Treinamento do Modelo

````python
historico = modelo.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    verbose=1,
    shuffle=False
)

print(" Treinamento concluído com sucesso")
#  TREINAMENTO: 50 épocas com validação em tempo real
````

##AVALIAÇÃO DE PERFORMANCE

````python
# previsões
previsoes_treino = modelo.predict(X_train)
previsoes_teste = modelo.predict(X_test)

# Convertendo para valores reais
previsoes_treino_reais = scaler.inverse_transform(previsoes_treino)
previsoes_teste_reais = scaler.inverse_transform(previsoes_teste)
y_treino_reais = scaler.inverse_transform(y_train.reshape(-1, 1))
y_teste_reais = scaler.inverse_transform(y_test.reshape(-1, 1))
````

````python
# Calculando métricas
metricas = {
    'MAE Treino': mean_absolute_error(y_treino_reais, previsoes_treino_reais),
    'MAE Teste': mean_absolute_error(y_teste_reais, previsoes_teste_reais),
    'MAPE Treino': mean_absolute_percentage_error(y_treino_reais, previsoes_treino_reais) * 100,
    'MAPE Teste': mean_absolute_percentage_error(y_teste_reais, previsoes_teste_reais) * 100
}

print(" RESULTADOS DO MODELO:")
for metrica, valor in metricas.items():
    if 'MAPE' in metrica:
        print(f"   {metrica}: {valor:.2f}%")
    else:
        print(f"   {metrica}: ${valor:.2f}")
````

````python
# Análise de overfitting
ratio_overfitting = metricas['MAE Treino'] / metricas['MAE Teste']
print(f"   Razão Treino/Teste: {ratio_overfitting:.2f}")
````

- MAPE de 4.15% indica ótima performance

##SISTEMA DE PREVISÃO

````python
def prever_futuro(dias_previsao=30):
    """
    Sistema profissional de previsão de preços
    """
    # Usar os últimos dados disponíveis
    dados_recentes = test_data[-TIME_STEPS:]

    previsoes = []
    sequencia_atual = dados_recentes.copy()

    for _ in range(dias_previsao):
        # Preparar entrada
        entrada = sequencia_atual.reshape(1, TIME_STEPS, 1)

        # Fazer previsão
        previsao = modelo.predict(entrada, verbose=0)[0, 0]
        previsoes.append(previsao)

        # Atualizar sequência
        sequencia_atual = np.roll(sequencia_atual, -1)
        sequencia_atual[-1] = previsao

    # Converter para valores reais
    previsoes_reais = scaler.inverse_transform(np.array(previsoes).reshape(-1, 1))

    return previsoes_reais.flatten()

# Gerar previsões
print("PREVISÕES FUTURAS:")
horizontes = [7, 30, 90]  # 1 semana, 1 mês, 3 meses
````

````python
for dias in horizontes:
    previsoes = prever_futuro(dias)
    variacao_percentual = ((previsoes[-1] - previsoes[0]) / previsoes[0]) * 100

    print(f"\n  Período: {dias} dias")
    print(f"   Preço inicial: ${previsoes[0]:.2f}")
    print(f"   Preço final: ${previsoes[-1]:.2f}")
    print(f"   Variação: {variacao_percentual:+.1f}%")
    print(f"   Faixa: ${previsoes.min():.2f} - ${previsoes.max():.2f}")
    #Sistema flexível para múltiplos horizontes
````

````python
def analise_aplicacao_pratica():
    """
    Análise do impacto em negócios de combustível
    """
    # Parâmetros realistas
    consumo_diario = 15  # barris/dia (posto médio)
    preco_atual = df_clean['Petrol'].iloc[-1]

    # Previsão para 30 dias
    previsao_30 = prever_futuro(30)
    preco_medio_previsto = np.mean(previsao_30)

    # Cálculos de impacto
    diferenca_preco = preco_medio_previsto - preco_atual
    impacto_mensal = diferenca_preco * consumo_diario * 30

    print("\n ANÁLISE PARA GESTÃO DE COMBUSTÍVEL:")
    print("=" * 45)
    print(f"Cenário Atual:")
    print(f"   • Preço atual: ${preco_atual:.2f}/barril")
    print(f"   • Consumo diário: {consumo_diario} barris")

    print(f"\n Previsão 30 Dias:")
    print(f"   • Preço médio previsto: ${preco_medio_previsto:.2f}")
    print(f"   • Tendência: {'ALTA' if diferenca_preco > 0 else 'BAIXA'}")

    print(f"\n Impacto Financeiro:")
    if impacto_mensal > 0:
        print(f"   • Aumento de custos: ${impacto_mensal:,.0f}/mês")
        print(f"   • Recomendação: Antecipar compras")
    else:
        print(f"   • Economia potencial: ${abs(impacto_mensal):,.0f}/mês")
        print(f"   • Recomendação: Postergar compras")

    # Estratégia baseada na previsão
    if abs(diferenca_preco) > 2.0:
        estrategia = "AJUSTAR ESTRATÉGIA DE COMPRAS"
    else:
        estrategia = "MANTER ESTRATÉGIA ATUAL"

    print(f"\nESTRATÉGIA: {estrategia}")

analise_aplicacao_pratica()
# APLICAÇÃO: Análise concreta para tomada de decisão
````
